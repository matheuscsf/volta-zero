# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
# Definindo os parmetros
media = 15  # mdia em kg
desvio_padrao = 0.1  # desvio padro em kg
peso = 15.2  # valor buscado
```

```python
# Calculando o valor Z
z = (peso - media) / desvio_padrao
```

```python
# Calculando a probabilidade de o pacote ter mais que 15,2 kg
probabilidade = stats.norm.sf(z)

probabilidade
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Calculando o percentil de ordem 95%
percentil_95 = stats.norm.ppf(0.95, loc=media, scale=desvio_padrao)

percentil_95
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu cdigo comea aqui

# Calculando as probabilidades para Y1 (Normal Padro)
probabilidades_Y1 = stats.norm.cdf(Y)

# Calculando as probabilidades para Y2 (t-Student com 5 graus de liberdade)
probabilidades_Y2 = stats.t.cdf(Y, df=5)
```

```python
probabilidades_Y1
```

```python
probabilidades_Y2
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
# Criando o grfico
plt.figure(figsize=(10, 6))
plt.plot(Y, probabilidades_Y1, label='Normal Padro (Y1)', color='blue')
plt.plot(Y, probabilidades_Y2, label='t-Student (Y2) com 5 graus de liberdade', color='red')
plt.title('Distribuies Cumulativas para Y1 e Y2')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade Cumulativa')
plt.legend()
plt.grid(True)
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
P=np.linspace(0, 1, 20)
# Seu cdigo comea aqui

# Calculando os percentis para Y1 (Normal Padro)
percentis_Y1 = stats.norm.ppf(P)

# Calculando os percentis para Y2 (t-Student com 5 graus de liberdade)
percentis_Y2 = stats.t.ppf(P, df=5)
```

```python
percentis_Y1
```

```python
percentis_Y2
```

6. Faa um grfico com os dados do item anterior.

```python
# Criando o grfico
plt.figure(figsize=(10, 6))
plt.plot(P, percentis_Y1, label='Normal Padro (Y1)', color='blue', marker='o')
plt.plot(P, percentis_Y2, label='t-Student (Y2) com 5 graus de liberdade', color='red', marker='x')
plt.title('Percentis para Y1 e Y2')
plt.xlabel('Probabilidade (P)')
plt.ylabel('Valores dos Percentis')
plt.legend()
plt.grid(True)
plt.show()
```
