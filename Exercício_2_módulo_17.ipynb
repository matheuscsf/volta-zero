````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
````

````python
# Carregar os dados de treino e teste
caminho_treino_x = 'X_train.csv'
dados_treino_x  = pd.read_csv(caminho_treino_x)
dados_treino_x.head()
````

````python
# Carregar os dados de treino e teste
caminho_teste_x = 'X_test.csv'
dados_teste_x  = pd.read_csv(caminho_teste_x)
dados_teste_x.head()
````

````python
# Carregar os rótulos de treino e teste
caminho_treino_y = 'y_train.txt'
dados_treino_y = pd.read_csv(caminho_treino_y, header=None, names=["label"])
dados_treino_y.head()
````

````python
# Carregar os rótulos de treino e teste
caminho_teste_y = 'y_test.txt'
dados_teste_y = pd.read_csv(caminho_teste_y, header=None, names=["label"])
dados_teste_y.head()
````

### 2. Divisão da base em Treino, Validação e Teste

A base já se encontra dividida em Treino e Validação. O que vamos fazer então é extrair uma base de Validação da base de Treino.

Extraia 25% da base de treino como base de validação.

````python
# Dividindo a base de treino em subconjuntos de treino e validação
dados_treino_x_final, dados_validacao_x, dados_treino_y_final, dados_validacao_y = train_test_split(
    dados_treino_x, 
    dados_treino_y, 
    test_size=0.25, 
    random_state=42
)

# Verificando o tamanho das novas bases de dados
(dados_treino_x_final.shape, dados_validacao_x.shape, dados_treino_y_final.shape, dados_validacao_y.shape)
````

### 3. Melhores 3 variáveis

Rode uma árvore com profundidade máxima igual a 4 para prever a atividade humana com todas as variáveis.
Observe a importância das variáveis e considere as 3 variáveis com maior importância para os próximos passos.
Dica: utilize o atributo ```clf.feature_importances_``` da árvore treinada.

````python
# Criando o modelo de árvore de decisão com profundidade máxima de 4
modelo_arvore = DecisionTreeClassifier(max_depth=4, random_state=42)

# Treinando o modelo com o conjunto de treino final
modelo_arvore.fit(dados_treino_x_final, dados_treino_y_final)

# Extraindo a importância das variáveis
importancias = modelo_arvore.feature_importances_

# Identificando as 3 variáveis mais importantes
indices_importantes = importancias.argsort()[-3:][::-1]
nomes_variaveis_importantes = dados_treino_x.columns[indices_importantes]
importancias_variaveis_importantes = importancias[indices_importantes]

# Exibindo as 3 variáveis mais importantes
nomes_variaveis_importantes, importancias_variaveis_importantes
````

### 4. Construa uma árvore com as 3 melhores variáveis

Utilizando as três variáveis encontradas acima, construa uma árvore de decisão. Encontre o melhor ```ccp_alpha``` utilizando a base de validação, conforme a estrutura que vimos em aula.

````python
%%time
# Selecionando apenas as 3 variáveis mais importantes para treino e validação
X_treino_reduzido = dados_treino_x_final[nomes_variaveis_importantes]
X_validacao_reduzido = dados_validacao_x[nomes_variaveis_importantes]

# Treinando o modelo de árvore de decisão sem poda
modelo_arvore_reduzido = DecisionTreeClassifier(random_state=42)
modelo_arvore_reduzido.fit(X_treino_reduzido, dados_treino_y_final)

# Calculando a complexidade e o erro de classificação para a base de validação
path = modelo_arvore_reduzido.cost_complexity_pruning_path(X_treino_reduzido, dados_treino_y_final)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

# Testando diferentes valores de ccp_alpha e armazenando a acurácia para cada um
acc_scores = []
for ccp_alpha in ccp_alphas:
    modelo = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    modelo.fit(X_treino_reduzido, dados_treino_y_final)
    y_pred = modelo.predict(X_validacao_reduzido)
    accuracy = accuracy_score(dados_validacao_y, y_pred)
    acc_scores.append(accuracy)

# Encontrando o ccp_alpha com a melhor acurácia
melhor_alpha = ccp_alphas[acc_scores.index(max(acc_scores))]
melhor_alpha, max(acc_scores)
````

### 5. Avaliação do modelo

Avalie a árvore encontrada no item anterior na base de testes.

````python
# Criando e treinando a árvore de decisão com o melhor ccp_alpha encontrado
modelo_final = DecisionTreeClassifier(random_state=42, ccp_alpha=melhor_alpha)
modelo_final.fit(X_treino_reduzido, dados_treino_y_final)

# Selecionando apenas as 3 variáveis mais importantes da base de teste
X_teste_reduzido = dados_teste_x[nomes_variaveis_importantes]

# Fazendo previsões na base de teste
y_pred_teste = modelo_final.predict(X_teste_reduzido)

# Calculando a acurácia na base de teste
acuracia_teste = accuracy_score(dados_teste_y, y_pred_teste)
acuracia_teste
````

A acurácia da árvore de decisão na base de testes é de aproximadamente 68.85%
