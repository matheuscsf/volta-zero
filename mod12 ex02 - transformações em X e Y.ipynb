# Regressão 01 - tarefa 02 - transformações em X e Y

### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

````python
import pandas as pd
import seaborn as sns
from seaborn import load_dataset

import matplotlib.pyplot as plt

import numpy as np

import statsmodels.formula.api as smf
import statsmodels.api as sm

from sklearn.preprocessing import PolynomialFeatures
````

````python
tips_df = sns.load_dataset("tips")
tips_df['tip_pct'] = tips_df['tip'] / (tips_df['total_bill'] - tips_df['tip'])
tips_df['net_bill'] = tips_df['total_bill'] - tips_df['tip']
tips_df.head()
````

````python
X_pct = tips_df['net_bill']
y_pct = tips_df['tip_pct']
X_pct = sm.add_constant(X_pct)

model_pct = sm.OLS(y_pct, X_pct).fit()

model_pct_summary = model_pct.summary()
model_pct_summary
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

````python
import numpy as np

# Preparando os dados para a regressão com o logaritmo de 'net_bill'
X_log = np.log(tips_df['net_bill'])  # aplicando logaritmo
y = tips_df['tip_pct']               # variável dependente

# Adicionando a constante para o intercepto
X_log = sm.add_constant(X_log)

# Criando o modelo de regressão linear com logaritmo
model_log = sm.OLS(y, X_log)

# Ajustando o modelo
results_log = model_log.fit()

# Exibindo os resultados
results_log.summary()
````

R-quadrado: 0.202. Isso significa que cerca de 20.2% da variabilidade em tip_pct é explicada pelo logaritmo de net_bill. Esse modelo apresenta uma melhoria em relação ao modelo anterior sem logaritmo.

Intercepto (constante): 0.6181. Indica que, se o logaritmo de net_bill fosse 0, o percentual esperado da gorjeta seria aproximadamente 61.81%.

Coeficiente de net_bill (logarítmico): -0.1534. Este coeficiente negativo sugere que, à medida que o valor de net_bill aumenta, o percentual da gorjeta diminui, mesmo após a transformação logarítmica.

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.

````python
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(tips_df[['net_bill']])

X_poly = sm.add_constant(X_poly)

model_poly = sm.OLS(y, X_poly).fit()

model_poly.summary()
````

R-quadrado representa uma melhoria em relação ao modelo linear simples, indicando que aproximadamente 15.8% da variabilidade em 'tip_pct' é explicada pelo modelo polinomial de 'net_bill'.

###  4. ajuste um modelo no log do ```tip_pct``` em função de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```

````python
y_log = np.log(tips_df['tip_pct'])  # Transformação logarítmica
X = tips_df[['net_bill']]           # Variável independente
X = sm.add_constant(X)              # Adicionar uma constante ao modelo

model_log = sm.OLS(y_log, X).fit()

y_pred_log = model_log.predict(X)
y_pred = np.exp(y_pred_log)  # Aplicar a transformação inversa do logaritmo

# Calcular R-quadrado manualmente para a transformação inversa
ss_res = np.sum((tips_df['tip_pct'] - y_pred) ** 2)
ss_tot = np.sum((tips_df['tip_pct'] - np.mean(tips_df['tip_pct'])) ** 2)
r_squared = 1 - (ss_res / ss_tot)

model_log.summary(), r_squared
````

###  5. ajuste um modelo no log do ```tip_pct``` em função do log de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```.

````python
X_log = np.log(tips_df['net_bill'])  
X_log = sm.add_constant(X_log)    

model_log_log = sm.OLS(y_log, X_log).fit()

y_pred_log_log = model_log_log.predict(X_log)
y_pred_log = np.exp(y_pred_log_log)  # Aplicar a transformação inversa do logaritmo

ss_res_log = np.sum((tips_df['tip_pct'] - y_pred_log) ** 2)
ss_tot_log = np.sum((tips_df['tip_pct'] - np.mean(tips_df['tip_pct'])) ** 2)
r_squared_log = 1 - (ss_res_log / ss_tot_log)

model_log_log.summary(), r_squared_log
````

###  6. qual modelo resultou no melhor ajuste?

Entre os modelos ajustados, o modelo que utilizou o logaritmo de 'tip_pct' em função do logaritmo de 'net_bill' resultou no melhor ajuste.

O modelo Log-Log não só apresentou o maior R-squared (0.293) para os dados na escala logarítmica, como também, ao converter de volta para a escala original de 'tip_pct', manteve um R-squared relativamente alto (0.228).
