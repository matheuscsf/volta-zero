####  2) Faça um boxplot do peso do bebê

```python
#Seaborn

sns.boxplot(y="PESO", data=df)
```

####  3) Faça um histograma da idade da mãe

```python
#Matplotlib

df['IDADEMAE'].plot.hist()
```

```python
#Seaborn

sns.histplot(df['IDADEMAE'], kde=True, color='skyblue', element = 'step')
plt.title('Histograma da idade da mãe')
plt.xlabel('Idade da mãe')
plt.ylabel('Frequência')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
#Seaborn

sns.histplot(df['PESO'], kde=True, color='skyblue', element = 'step')
plt.title('Histograma do Peso')
plt.xlabel('Peso')
plt.ylabel('Frequência')
plt.show()
```
